**Not all activities were compulsory.**

The column names that end with "kokku" are category totals, all columns preceding
it until the previous "kokku" belong to that category.

The column "Kogutulemus (Punktid)" is the student's final result on a scale of 0-110. The column "Hinne (Punktid)" is the result according to which the grade is determined. The grading scale was A [90-110], B [80-90), C [70-80), D [60-70), E [50-60), F [0-50).

```python
grades_en.drop(['Grade (letter)'], axis=1, inplace=True)
#grades_en.columns
```

```python
grades_en_cpy = grades_en.copy()
```

```python
rows = len(grades_en)
test = np.ones(342)

grades_en_cpy.insert(len(grades_en_cpy.columns), "Grade (letter)", (test))

for row in range(0, rows):
    grade = int(grades_en_cpy['Grade (points) '].loc[row])

    if grade < 50:
        grades_en_cpy['Grade (letter)'].loc[row] = 'F'
    
    elif (grade >= 50) and (grade <= 59):
        grades_en_cpy['Grade (letter)'].loc[row] = 'E'
    
    elif (grade >= 60) and (grade <= 69):
        grades_en_cpy['Grade (letter)'].loc[row] = 'D'
    
    elif (grade >= 70) and (grade <= 79):
        grades_en_cpy['Grade (letter)'].loc[row] = 'C'
    
    elif (grade >= 80) and (grade <= 89):
        grades_en_cpy['Grade (letter)'].loc[row] = 'B'
    
    elif (grade >= 90) and (grade <= 100):
        grades_en_cpy['Grade (letter)'].loc[row] = 'A'
```

```python

import time

start_time = time.time()

logbook_Inf_02 = pd.read_excel('logs_LTAT.03.001_20210207-0942_Inf_02.xlsx') # read in logbook

read_time = time.time()
print(f"reading the logbook in took this long {round(read_time - start_time)}s")

# translate columns' name using rename function
logbook_Inf_02.rename(columns=lambda x: translator.translate(x, lang_src='et', lang_tgt='en'), inplace=True)    
translation_time = time.time()

print(f"translating {len(logbook_Inf_02.columns)} column names took this long {round(translation_time - start_time)}s")
```

```python
#logbook_Uld_02_translated['Component'] = logbook_Uld_02['Komponent'].apply(lambda x: translator.translate(x, lang_src='et', lang_tgt='en'))
#print(logbook_Inf_02.columns)
from multiprocessing.dummy import Pool as ThreadPool

pool = ThreadPool(4) # nm of threads possible

def request(text):
    t = google_translator(timeout=10)
    translate_text = t.translate(text, "en")
    return translate_text

start = time.time()

values_event_context = logbook_Inf_02['Event context '].value_counts().index.tolist() # selects values in each column and returns value counts
values_name_of_event = logbook_Inf_02['Name of event '].value_counts().index.tolist() # selects values in each column and returns value counts
values_component = logbook_Inf_02['Component '].value_counts().index.tolist() # selects values in each column and returns value counts

values_read_time = time.time()
            
print(f"reading in values took this long: {round(values_read_time - start)}s")

try:
    results_component = pool.map(request, values_component)
    results_name_of_event = pool.map(request, values_name_of_event)
    results_event_context = pool.map(request, values_event_context)

except Exception as e:
    raise e
finally:
    print(f"translating took this long: {round(time.time() - start)}s")
    pool.close()
    pool.join()

# substitute og with transl vals
```

```python
logbook_Inf_02_cpy = logbook_Inf_02.copy()
start_time = time.time()
for i in range(0, len(logbook_Inf_02_cpy)):
    value_component = logbook_Inf_02_cpy['Component '].loc[i]
    x=0
    for val in values_component:
        if value_component == val:
            logbook_Inf_02_cpy['Component '].loc[i] = results_component[x]
            #print("Replaced with translation.")
        x+=1
end_time = time.time()
print(f"\nprocessing took {round(end_time - start_time)}s")
```
